```python

# Get unique categorical values for each column
unique_values = {}
for column in df.columns:
    if df[column].dtype == 'O':  # 'O' represents object data type (categorical)
        unique_values[column] = df[column].unique()

# Display unique values for each categorical column
print("\nUnique Categorical Values for Each Column:")
for column, values in unique_values.items():
    print(f"{column}: {values}")
```

```python

# Display the original DataFrame
print("Original DataFrame:")
print(df)

# Create mapping dictionaries
manufacturer_mapping = {val: idx + 1 for idx, val in enumerate(df['manufacturer'].unique())}
condition_mapping = {val: idx + 1 for idx, val in enumerate(df['condition'].unique())}
cylinders_mapping = {val: idx + 1 for idx, val in enumerate(df['cylinders'].unique())}
fuel_mapping = {val: idx + 1 for idx, val in enumerate(df['fuel'].unique())}
transmission_mapping = {val: idx + 1 for idx, val in enumerate(df['transmission'].unique())}
drive_mapping = {val: idx + 1 for idx, val in enumerate(df['drive'].unique())}
size_mapping = {val: idx + 1 for idx, val in enumerate(df['size'].unique())}
type_mapping = {val: idx + 1 for idx, val in enumerate(df['type'].unique())}
paint_color_mapping = {val: idx + 1 for idx, val in enumerate(df['paint_color'].unique())}
state_mapping = {val: idx + 1 for idx, val in enumerate(df['state'].unique())}

# Apply mappings to the DataFrame
df['manufacturer'] = df['manufacturer'].map(manufacturer_mapping)
df['condition'] = df['condition'].map(condition_mapping)
df['cylinders'] = df['cylinders'].map(cylinders_mapping)
df['fuel'] = df['fuel'].map(fuel_mapping)
df['transmission'] = df['transmission'].map(transmission_mapping)
df['drive'] = df['drive'].map(drive_mapping)
df['size'] = df['size'].map(size_mapping)
df['type'] = df['type'].map(type_mapping)
df['paint_color'] = df['paint_color'].map(paint_color_mapping)
df['state'] = df['state'].map(state_mapping)

# Display the DataFrame after mapping
print("\nDataFrame after mapping:")
print(df)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split

from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.preprocessing import  StandardScaler,LabelEncoder,OneHotEncoder
```

```python
# Defining the features and the target of the model

X = df.drop(['price'], axis=1)     # Features
y = df.price                # Target 
```

```python
# Breaking the data into train and test subsets

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

## Linear Regression

```python
# Importing 'LinearRegression' through linear_model module

from sklearn.linear_model import LinearRegression  

reg = LinearRegression()         # Instantiate
reg.fit(X_train, y_train)        # Fit the train data

r2_train_score = reg.score(X_train, y_train)   # Calculating R^2 score for train

print('R^2 score for train dataset = ', round(r2_train_score, 4), '\n')
print('Coefficients of Linear Model:', reg.coef_, '\n')
print('Intercept:', reg.intercept_)
```

```python

# Finding the predictions of the model for test dataset

y_pred = reg.predict(X_test)

y_pred[:10]  # Representing the price prediction for the first 10 data points in test dataset
```

```python

# Evaluating the performance of the model on the test dataset

r2_test_score = reg.score(X_test, y_test)   # Calculating R^2 score for train

print('R^2 score for test dataset = ', round(r2_test_score, 4), '\n')
```

## Decision Tree

```python
# Constructing a tree regressor
tr_reg = DecisionTreeRegressor()
tr_reg.fit(X_train, y_train)
```

```python
# Finding the predictions of the tree regressor for train and test subsets
train_y_pred = tr_reg.predict(X_train)
test_y_pred = tr_reg.predict(X_test)
```

```python
r2_train_score = tr_reg.score(X_train, y_train) 
r2_test_score = tr_reg.score(X_test, y_test)
print('R^2 score for train dataset = ', round(r2_train_score, 4), '\n')
print('R^2 score for test dataset = ', round(r2_test_score, 4), '\n')
```

## Grid Search CV

```python
# Using 'GridSearchCV' to find the best hyperparameters

from sklearn.model_selection import GridSearchCV

params = {'max_leaf_nodes': list(range(2, 100)), 'min_samples_split': [2, 3, 4]}

grid_search_cv = GridSearchCV(DecisionTreeRegressor(random_state=3), params, verbose=1, cv=3)
grid_search_cv.fit(X_train, y_train)
```

```python
# Displaying the best parameters

grid_search_cv.best_estimator_
```

```python
# Retraining the tree regressor with the best hyperparameters

tr_reg = DecisionTreeRegressor(max_leaf_nodes=15, min_samples_split=4, random_state=3)
tr_reg.fit(X_train, y_train)
```

```python
# Finding the predictions of the latter tree regressor for train and test subsets

train_y_pred = tr_reg.predict(X_train)
test_y_pred = tr_reg.predict(X_test)
```

```python
r2_train_score = tr_reg.score(X_train, y_train)   # Calculating R^2 score for train
r2_test_score = tr_reg.score(X_test, y_test)      # Calculating R^2 score for test

print('R^2 score for train dataset = ', round(r2_train_score, 4), '\n')
print('R^2 score for test dataset = ', round(r2_test_score, 4), '\n')
```

## $\color{ForestGreen}{\text{Random Forest Regressor}}$

```python
# Constructing a random forest regressor

from sklearn.ensemble import RandomForestRegressor

rf_reg = RandomForestRegressor(n_estimators=1000, random_state=3)  # Instantiating the forest
rf_reg.fit(X_train, y_train)     
```

```python
# Finding the predictions of the random forest regressor for train and test subsets

train_y_pred = rf_reg.predict(X_train)
test_y_pred = rf_reg.predict(X_test)
```

```python
r2_train_score = rf_reg.score(X_train, y_train)   # Calculating R^2 score for train
r2_test_score = rf_reg.score(X_test, y_test)      # Calculating R^2 score for test

print('R^2 score for train dataset = ', round(r2_train_score, 4), '\n')
print('R^2 score for test dataset = ', round(r2_test_score, 4), '\n')
```

### Comparision of Models

```python
import matplotlib.pyplot as plt
import numpy as np

# Example R2 scores (replace these with your actual scores)
algorithms = ['Decision Tree','Grid Search CV', 'Random Forest']
train_scores = [0.998, 0.6007, 0.9796]  # R2 scores on the training set
test_scores = [0.722, 0.6095, 0.8566]  # R2 scores on the testing set

# Plotting
bar_width = 0.35
index = np.arange(len(algorithms))

plt.figure(figsize=(10, 5))

# Plotting the bars
train_bars = plt.bar(index, train_scores, bar_width, label='Train', alpha=0.8)
test_bars = plt.bar(index + bar_width, test_scores, bar_width, label='Test', alpha=0.8)

# Adding R2 score values above each bar
for bar, score in zip(train_bars, train_scores):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 0.05, f'{score:.3f}', ha='center', color='white', fontweight='bold')

for bar, score in zip(test_bars, test_scores):
    plt.text(bar.get_x() + bar.get_width() / 2, bar.get_height() - 0.05, f'{score:.3f}', ha='center', color='black', fontweight='bold')

# Adding labels and legend
plt.xlabel('Algorithms')
plt.ylabel('R2 Score')
plt.title('Comparison of R2 Scores on Training and Testing Sets')
plt.xticks(index + bar_width / 2, algorithms)
plt.legend()

# Adjusting layout and displaying the plot
plt.tight_layout()
plt.show()
```
